### 7.1 Create fam_status flags

```python
# use loof to replace 'divorced/widowed' and 'living with parents' to 'single'

result = []

for value in df["fam_status"]:
  if value == 'married':
    result.append("married")

  else:
    result.append("single")
```

```python
# Assign a new column in the df using the 'result' list above

df['Fam_status']=result
```

```python
df.head()
```

```python
# dropping the olf column of 'fam_status'

df.drop(columns='fam_status', inplace=True)
```

### 7.2 Split date value and create new colum

```python
## Trying to split the 'date_joined' into a new colum containing only 'year'

df['year_joined'] = df['date_joined'].str.split('/').str[-1]
```
