# 8 Create spending flags

Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# Use transform() to determine the mean price of products purchased by each user

df['mean_product_price'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# view head

df.head()
```

```python
# Use .loc() to assign which values in 'mean_product_price' will receive a certain value

df.loc[df['mean_product_price']<10, 'customer_typ']='Low spender'
df.loc[df['mean_product_price']>=10, 'customer_typ']='High spender'
```

```python
df.head()
```

# 7. Create column for frequency flags

Create a column for frequency flags for the customer, based on the median of their in the “days_since_prior_order” column.

Criteria:

If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”


If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”


If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”




```python
# Use groupby and transform to determine median number of days between orders for each user

df['median_days_between_orders'] = df.groupby(['user_id'])['days_since_previous_order'].transform(np.median)
```

```python
df[['days_since_previous_order','median_days_between_orders','user_id']].head(10)
```

```python
# Create order frequency flag, using the median_days_between_orders

df.loc[df['median_days_between_orders']>20, 'frequency_flag']='Non-frequent Customer'
df.loc[(df['median_days_between_orders']>10) & (df['median_days_between_orders']<=20), 'frequency_flag']='Regular-frequent Customer'
df.loc[df['median_days_between_orders']<=10, 'frequency_flag']='Frequent Customer'
```

```python
# Check for head

df[['median_days_between_orders','user_id','frequency_flag']].head(10)
```

```python
# Check value counts of new order_frequency column

df['frequency_flag'].value_counts(dropna = False)
```
